# Create Dataset

```python
import os
import pandas as pd

columns = ['image_name', 'upper_boundary', 'lower_boundary', 'mean_thickness', 'variance_thickness']

image_dir = 'SCL_OCT_dataset/label/'  
image_files = [f for f in os.listdir(image_dir) if f.endswith('.png')]
```

```python
all_cornea_data = []

for image_file in image_files:
    image_path = os.path.join(image_dir, image_file)
    label_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    upper_boundary_value = 102 
    lower_boundary_value = 153 

    upper_boundary = np.argmax(label_image >= upper_boundary_value, axis=0)

    lower_boundary = label_image.shape[0] - np.argmax(np.flipud(label_image) < lower_boundary_value, axis=0)
    thickness = lower_boundary - upper_boundary

    mean_thickness = np.mean(thickness)
    variance_thickness = np.var(thickness)

    cornea_data = [image_file, upper_boundary, lower_boundary, mean_thickness, variance_thickness]
    all_cornea_data.append(cornea_data)
```

```python
results_df = pd.DataFrame(all_cornea_data, columns=columns)
```

```python
results_df
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


n_clusters = 3 

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
results_df['cluster'] = kmeans.fit_predict(results_df[['mean_thickness']])

plt.scatter(results_df['image_name'], results_df['mean_thickness'], c=results_df['cluster'], cmap='viridis')
plt.title('Corneal Thickness Clusters')
plt.xlabel('Image Name')
plt.ylabel('Mean Thickness (pixels)')
plt.xticks(rotation=90)
plt.colorbar(label='Cluster')
plt.show()
```
